## Topics:

- Matplotlib: It is one of the data science packages.
    - It is a visualizatoin package.
    - We can represent the data in the form of graphs or pictures.
    
- How can we import matplotlib:
    - import matplotlib.pyplot as plt
    - from matplotlib import pyplot as plt
    
    
- Types of Plots:
    - Line plot
        - linear
        - non-linear
    - Bar plot
    - Histrogram
    - pie chart
    - scatter plot etc..

```python
# (3,8) and (10,15)
from matplotlib import pyplot as plt

x = [3,10] # x points
y = [8,15] # y points

plt.plot(x,y)

plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]

plt.plot(x1,y1)

plt.xlabel("X axis") # x axis name
plt.ylabel("Y axis") # y axis name

plt.title("Line Plot")

plt.show()
```

```python
a = [1,5,2,4,6]
b = [30,80,45,60,100]

p = [2,5,8,7,3]
q = [20,50,80,70,30]

plt.plot(a, b, label = "James")
plt.plot(p, q, label = "John")

plt.xlabel("Time (Hours)") # x axis name
plt.ylabel("Distance (KM)") # y axis name

plt.legend()

plt.title("Time vs Distance",loc = "right")

plt.show()
```

```python
import numpy as np

x = np.arange(5) # 0,1,2,3,4
y = np.arange(5,10) # 5,6,7,8,9

plt.plot(x,y,linestyle = "dotted",lw = 5,c= "cyan")

plt.show()
```

### Barplot

```python
# 50 students and grades
student_names = ["ram","syam","viplav","vijay","sandeep","rasshu"]
attendance = [10,4,15,1,8,13]

plt.bar(student_names,attendance,color="violet",edgecolor="green",lw = 2,width = 0.3) # default 0.8

plt.show()
```

```python
student_names = ["ram","syam","viplav","vijay","sandeep","rasshu"]
attendance = [10,4,15,1,8,13]

plt.barh(student_names,attendance,color="violet",edgecolor="green",lw = 2,height=0.6)# default 0.8

plt.show()
```

## Histogram
- to represent the frequnecy

```python
x = [1,2,3,4,1,2,3,2,1,2,1,2,1,5,6] # ones-5,two - 5,three-2, four-1,five-1,six-1

plt.hist(x)

plt.show()
```

```python
 clrs = ['red','yellow','brown','hotpink','green','k']
plt.pie(attendance,labels = student_names,colors = clrs)
plt.show()
```
